```python
# üìä An√°lisis de Rendimiento de Estudiantes
# üë©‚Äçüíª Autor: Agustina Bel√©n Irisarri
# üìÅ Proyecto: Analisis_rendimiento_estudiantes

# ‚úÖ Importar librer√≠as
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ‚úÖ Cambiar directorio de trabajo al proyecto
os.chdir('C:/Users/agus1/OneDrive/Escritorio/Personal Agus/Analisis de Datos/Analisis_rendimiento_estudiantes')

# ‚úÖ Configuraci√≥n para mostrar gr√°ficos en el notebook
%matplotlib inline
sns.set(style="whitegrid")

# üìÇ Cargar el dataset
data = pd.read_csv('data/students_performance.csv')

# üîç Mostrar las primeras filas
print("üìå Primeras filas del dataset:")
display(data.head())

# ‚ÑπÔ∏è Informaci√≥n general del dataset
print("\n‚ÑπÔ∏è Informaci√≥n del dataset:")
display(data.info())

# üìä Estad√≠sticas descriptivas
print("\nüìä Estad√≠sticas descriptivas:")
display(data.describe())

# üìà Visualizaci√≥n: Distribuci√≥n de puntajes en matem√°ticas
plt.figure(figsize=(10,6))
sns.histplot(data['math_score'], bins=20, kde=True, color='blue')
plt.title('Distribuci√≥n de Puntajes en Matem√°ticas')
plt.xlabel('Puntaje')
plt.ylabel('Cantidad de Estudiantes')
plt.show()

# üìä Visualizaci√≥n: Boxplot de puntajes por g√©nero
plt.figure(figsize=(8,6))
sns.boxplot(x='gender', y='math_score', data=data, palette='Set2')
plt.title('Puntajes en Matem√°ticas por G√©nero')
plt.xlabel('G√©nero')
plt.ylabel('Puntaje en Matem√°ticas')
plt.show()

# üìä Relaci√≥n entre lectura y escritura
plt.figure(figsize=(8,6))
sns.scatterplot(x='reading_score', y='writing_score', hue='gender', data=data)
plt.title('Relaci√≥n entre Puntajes de Lectura y Escritura')
plt.xlabel('Puntaje de Lectura')
plt.ylabel('Puntaje de Escritura')
plt.legend(title='G√©nero')
plt.show()


```
